```python
np_exp = np.array(exp)
```

```python
from scipy.spatial import distance
```

```python
dist = distance.pdist(np_exp.T)
```

```python
import itertools
```

```python
distance_mat = distance.squareform(dist)
```

```python
from scipy.sparse.csgraph import minimum_spanning_tree
from scipy.sparse import csr_matrix, coo_matrix, lil_matrix
```

```python
mst=minimum_spanning_tree(distance_mat)
```

```python
edge = mst.toarray()
#  (i,j)または(j,i)にしか示されていないので変更
edge = np.where(edge>edge.T, edge, edge.T)

pd.DataFrame(edge).to_csv("mst_edge.csv")
```

edgeとrootを与えられたときにそれぞれのnodeまでの距離を巾優先探索で計算する

```python
from collections import deque
```

```python
start=time[1].values.argmin()
```

```python
pseudotime=np.repeat(-1.0, exp.shape[1])

parent=np.repeat(-1, exp.shape[1])
```

```python
def get_child(i, parent=None):
    connected = np.where(edge[i]!=0)[0]
    child=np.delete(connected, np.where(connected==parent))
    return child
```

```python
q=deque()
q.append(start)
pseudotime[start] = 0
parent[get_child(start)]=start
```

```python
while len(q)>0:
    node = q.popleft()
    if node == start:
        pseudotime[node]=0
        node_par=None
    else:
        node_par=parent[node]
        parent_pseudo_time = pseudotime[node_par]
        pseudotime[node] = parent_pseudo_time + edge[node, node_par]
        #print("-----------")
        #print(parent_pseudo_time)
        #print(edge[node,node_par])
        #print(pseudotime[node])
    #print(pseudotime[node])
    #print(pseudotime)
    child=get_child(node,node_par)
    parent[child]=node
    q.extend(child)
```

```python
pseudotime_norm = pseudotime/np.max(pseudotime)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(exp_dr[:,0], exp_dr[:,1],s=2,c=pseudotime_norm, cmap="plasma")
plt.colorbar()
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("pseudotime by MST")
plt.savefig("pseudotime_mst.png",dpi=100)
```

```python
corrcoef = np.corrcoef(time.loc[:,1],pseudotime_norm)
corrcoef
```
